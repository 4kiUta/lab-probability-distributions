```python
# your code here
from scipy.stats import expon


def exp_dist_list(x, size = 10):
    return np.random.exponential(x, size)




seq_mena_1 = exp_dist_list(1,1000)
seq_mena_100 = exp_dist_list(100,1000)


fig, axs = plt.subplots(1,2, figsize=(20, 10)) 


axs[0].hist(seq_mena_1, bins = 100)
axs[1].hist(seq_mena_100, bins = 100)


```
